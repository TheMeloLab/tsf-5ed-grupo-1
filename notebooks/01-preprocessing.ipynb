```python
import pandas as pd
```

```python
# Load dataset
df = pd.read_csv("../data/raw/dataset.csv")
```

```python
df["DateUTC"] = pd.to_datetime(df["DateUTC"])

# Get latest date in dataset
max_date = df["DateUTC"].max()

# Last 3 years
cutoff_date = max_date - pd.DateOffset(years=3)

df_recent = df[df["DateUTC"] >= cutoff_date].copy()
```

```python
df_iberia = df_recent[df_recent["CountryCode"].isin(["PT", "ES"])].copy()
df_pt = df_iberia[df_iberia["CountryCode"] == "PT"].copy()
df_es = df_iberia[df_iberia["CountryCode"] == "ES"].copy()
```

```python
cols = ["DateUTC", "CountryCode", "Value", "Cov_ratio"]

df_pt = df_pt.reset_index()
df_es = df_es.reset_index()
df_iberia = df_iberia.reset_index()

df_pt = df_pt[cols]
df_es = df_es[cols]
df_iberia = df_iberia[cols]
```

```python
df_pt = df_pt.set_index("DateUTC").sort_index()
df_es = df_es.set_index("DateUTC").sort_index()
df_iberia = df_iberia.set_index("DateUTC").sort_index()
```

```python
df_pt.to_csv("../data/processed/portugal.csv")
df_es.to_csv("../data/processed/spain.csv")
df_iberia.to_csv("../data/processed/iberia.csv")
```
